```python
from folium.plugins import MarkerCluster


# Calculate mean 'class' value for each launch site
mean_class = spacex_df.groupby('Launch Site')['class'].mean().reset_index()
mean_class.columns = ['Launch Site', 'mean_class']

# Merge mean_class with spacex_df
launch_site = spacex_df.merge(mean_class, on='Launch Site')

# Define function to determine color based on mean 'class' value
def get_color(value):
    if value < 0.33:
        return 'red'
    elif value < 0.67:
        return 'orange'
    else:
        return 'green'

# Initialize the map
nasa_coordinate = [29.559684, -95.083097]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Create a marker cluster object
marker_cluster = MarkerCluster()

# Add circles based on mean 'class' value
for index, site in mean_class.iterrows():
    color = get_color(site['mean_class'])
    circle = folium.Circle(
        location=[spacex_df[spacex_df['Launch Site'] == site['Launch Site']]['Lat'].values[0], 
                  spacex_df[spacex_df['Launch Site'] == site['Launch Site']]['Long'].values[0]],
        radius=100,
        color=color,
        fill=True,
        fill_color=color,
        popup=folium.Popup(site['Launch Site'])
    ).add_to(site_map)

# Add markers based on actual 'class' value
for index, record in spacex_df.iterrows():
    marker_color = 'green' if record['class'] == 1 else 'red'
    
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=marker_color),
        popup=folium.Popup(f"{record['Launch Site']}<br>Class: {record['class']}")
    )
    
    marker_cluster.add_child(marker)

# Add marker cluster to the map
site_map.add_child(marker_cluster)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import radians, sin, cos, sqrt, atan2

formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Sample proximity data (latitude and longitude)
proximity_data = {
    'Proximity': ['City A', 'City B', 'City C'],
    'Lat': [29.7604, 28.5383, 27.9942],
    'Long': [-95.3698, -81.3792, -82.4562]
}
proximity_df = pd.DataFrame(proximity_data)

# Define the Haversine formula to calculate distance
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in kilometers
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2) * sin(dlat/2) + cos(lat1) * cos(lat2) * sin(dlon/2) * sin(dlon/2)
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance

# Calculate distances and add to a new dataframe
distances = []
for _, launch in launch_sites_df.iterrows():
    for _, prox in proximity_df.iterrows():
        distance = haversine(launch['Lat'], launch['Long'], prox['Lat'], prox['Long'])
        distances.append([launch['Launch Site'], prox['Proximity'], distance])

distances_df = pd.DataFrame(distances, columns=['Launch Site', 'Proximity', 'Distance (km)'])

# Display distances
print(distances_df)

from folium.features import DivIcon

# Initialize the map centered around the US
site_map = folium.Map(location=[29.559684, -95.083097], zoom_start=5)

# Add launch sites to the map
for _, site in launch_sites_df.iterrows():
    folium.Marker(
        location=[site['Lat'], site['Long']],
        popup=site['Launch Site'],
        icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
    ).add_to(site_map)

# Add proximity points to the map
for _, prox in proximity_df.iterrows():
    folium.Marker(
        location=[prox['Lat'], prox['Long']],
        popup=prox['Proximity'],
        icon=folium.Icon(color='purple')
    ).add_to(site_map)

# Add lines between launch sites and proximity points
for _, row in distances_df.iterrows():
    launch_site = launch_sites_df[launch_sites_df['Launch Site'] == row['Launch Site']].iloc[0]
    proximity_point = proximity_df[proximity_df['Proximity'] == row['Proximity']].iloc[0]
    
    folium.PolyLine(
        locations=[(launch_site['Lat'], launch_site['Long']), (proximity_point['Lat'], proximity_point['Long'])],
        color='blue'
    ).add_to(site_map)
    
    # Add distance text
    mid_lat = (launch_site['Lat'] + proximity_point['Lat']) / 2
    mid_long = (launch_site['Long'] + proximity_point['Long']) / 2
    folium.Marker(
        location=[mid_lat, mid_long],
        icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12pt; color: blue;">{row["Distance (km)"]:.2f} km</div>'
        )
    ).add_to(site_map)

# Display the map
site_map
```

```python
import folium
from folium.plugins import MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance between two points using the provided formula
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Initialize the map centered around the launch site
site_map = folium.Map(location=[28.5721, -80.648], zoom_start=10)

# Define the launch site coordinates
launch_site_lat = 28.5721
launch_site_lon = -80.648

# Add the launch site marker
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup='KSC LC-39A',
    icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
).add_to(site_map)

# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 7);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=7,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Proximity data for the launch site
ksclc39a_proximity_data = {
    'Proximity': ['City KSC LC-39A', 'Highway KSC LC-39A', 'Coastline KSC LC-39A'],
    'Lat': [28.6122181, 28.589958, 28.608490],
    'Long': [-80.8075534, -80.802699, -80.594398]
}

# Calculate and display distances
for i in range(len(ksclc39a_proximity_data['Proximity'])):
    proximity_name = ksclc39a_proximity_data['Proximity'][i]
    proximity_lat = ksclc39a_proximity_data['Lat'][i]
    proximity_lon = ksclc39a_proximity_data['Long'][i]
    
    distance = calculate_distance(launch_site_lat, launch_site_lon, proximity_lat, proximity_lon)
    
    # Add a marker on the proximity point with distance label
    distance_marker = folium.Marker(
        location=[proximity_lat, proximity_lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:.2f} KM".format(distance),
        )
    )
    site_map.add_child(distance_marker)
    
    # Create a PolyLine object to show the line between the launch site and the proximity
    lines = folium.PolyLine(locations=[[launch_site_lat, launch_site_lon], [proximity_lat, proximity_lon]], color='red', weight=1)
    site_map.add_child(lines)

# Display the map
site_map
```
